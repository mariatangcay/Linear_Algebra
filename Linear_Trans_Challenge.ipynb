2. Develop a python function from scratch that will find both the eigenvectors and eigenvalues of any $n \times n$ matrix.

3. Test your functions from a randomly generated $n \times n$ matrix.


```python
import numpy as np

# 1. Function to find the determinant of a matrix


def determinant(matrix):
    # 1.1. Check if matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError('The matrix is not square')

    # 1.2. Check if matrix is 2x2
    if len(matrix) == 2:
        return matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0]

    # 1.3. Check if matrix is larger than 2x2
    else:
        # 1.3.1. Initialize determinant
        det = 0

        # 1.3.2. Iterate through first row
        for i in range(len(matrix)):
            # 1.3.3. Initialize submatrix
            submatrix = []

            # 1.3.4. Iterate through rows
            for j in range(1, len(matrix)):
                # 1.3.5. Initialize subrow
                subrow = []

                # 1.3.6. Iterate through columns
                for k in range(len(matrix)):
                    # 1.3.7. Check if column is not equal to column of first row
                    if k != i:
                        # 1.3.8. Append element to subrow
                        subrow.append(matrix[j][k])

                # 1.3.9. Append subrow to submatrix
                submatrix.append(subrow)

            # 1.3.10. Add determinant of submatrix to determinant
            det += (-1)**i * matrix[0][i] * determinant(submatrix)

        # 1.3.11. Return determinant
        return det

# 2. Function to find the eigenvalues and eigenvectors of a matrix


def eigen(matrix):
    # 2.1. Check if matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError('The matrix is not square')

    # 2.2. Check if matrix is 2x2
    if len(matrix) == 2:
        # 2.2.1. Initialize eigenvalues
        eigenvalues = []

        # 2.2.2. Calculate eigenvalues
        eigenvalues.append((matrix[0][0] + matrix[1][1] + np.sqrt((matrix[0][0] + matrix[1][1])
                           ** 2 - 4*(matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0])))/2)
        eigenvalues.append((matrix[0][0] + matrix[1][1] - np.sqrt((matrix[0][0] + matrix[1][1])
                           ** 2 - 4*(matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0])))/2)

        # 2.2.3. Initialize eigenvectors
        eigenvectors = []

        # 2.2.4. Calculate eigenvectors
        eigenvectors.append([matrix[0][1], eigenvalues[0] - matrix[0][0]])
        eigenvectors.append([matrix[0][1], eigenvalues[1] - matrix[0][0]])

        # 2.2.5. Return eigenvalues and eigenvectors
        return eigenvalues, eigenvectors

    # 2.3. Check if matrix is larger than 2x2
    else:
        # 2.3.1. Initialize eigenvalues
        eigenvalues = []

        # 2.3.2. Calculate eigenvalues
        eigenvalues.append((matrix[0][0] + matrix[1][1] + np.sqrt((matrix[0][0] + matrix[1][1])
                           ** 2 - 4*(matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0])))/2)
        eigenvalues.append((matrix[0][0] + matrix[1][1] - np.sqrt((matrix[0][0] + matrix[1][1])
                           ** 2 - 4*(matrix[0][0]*matrix[1][1] - matrix[0][1]*matrix[1][0])))/2)

        # 2.3.3. Initialize eigenvectors
        eigenvectors = []

        # 2.3.4. Calculate eigenvectors
        eigenvectors.append([matrix[0][1], eigenvalues[0] - matrix[0][0]])
        eigenvectors.append([matrix[0][1], eigenvalues[1] - matrix[0][0]])

        # 2.3.5. Return eigenvalues and eigenvectors
        return eigenvalues, eigenvectors

# 3. Function to test a randomly generated nxn matrix


def test():
    # 3.1. Initialize matrix
    n = 2
    matrix = np.random.rand(n, n)

    # 3.2. Print matrix
    print('Matrix:')
    print(matrix)

    # 3.3. Print determinant
    print('Determinant:')
    print(determinant(matrix))

    # 3.4. Print eigenvalues and eigenvectors
    print('Eigenvalues and eigenvectors:')
    print(eigen(matrix))


# 4. Test function
test()
```
